```python
import matplotlib.pyplot as plt
x = np.linspace(0.8, 1.2)

plt.plot(x, np.sqrt(np.sin(np.pi*(21/180))**2 + np.cos(np.pi*(21/180))**2 * x**2 ) )
plt.plot(x, np.sqrt(np.sin(np.pi*(0/180))**2 + np.cos(np.pi*(0/180))**2 * x**2 ) )
plt.plot(x, np.sqrt(np.sin(np.pi*(90/180))**2 + np.cos(np.pi*(90/180))**2 * x**2 ) )
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, np.sqrt(np.sin(np.pi*(21/180))**2 + np.cos(np.pi*(21/180))**2 * x**2 )/np.sqrt(np.sin(np.pi*(0/180))**2 + np.cos(np.pi*(0/180))**2 * x**2 ) )
plt.ylabel('Error From Assuming Head-on', fontsize=16)
plt.ylim(0.98, 1.025)

plt.axvline(0.94), plt.axvline(1.06)
```
