```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

def reset_plt(ticksize,fontsize):
    plt.style.use('seaborn-white')
    plt.rcParams['xtick.labelsize'] = ticksize
    plt.rcParams['ytick.labelsize'] = ticksize
    plt.rcParams['font.size'] = fontsize
    #plt.rcParams['mathtext.fontset'] = 'stix'
    plt.rcParams['font.family'] = 'STIXGeneral'
    plt.rcParams['legend.facecolor'] = 'white'
    plt.rcParams['axes.formatter.limits'] = (-1,3)
    plt.rcParams['axes.linewidth'] = 2.25
    plt.rcParams['axes.grid'] = True
    plt.rcParams['grid.alpha'] = 0.6
```

```python
reset_plt(12, 20)

beta_lim = 0.99
x = np.linspace(-beta_lim, beta_lim, 1000)
y = np.linspace(-beta_lim, beta_lim, 1000)
X,Y = np.meshgrid(x,y)
beta = np.sqrt( X**2 + Y**2)#/(1-(X**2 + Y**2) )

plt.figure(figsize=(6,6), dpi=300)
plt.pcolormesh(X, Y, beta, cmap='Spectral'); 
im = plt.contour(X, Y, beta, [0.28, 0.33], cmap='Spectral'); 



#Classical
for beta_in in [0.28]:
    plt.axvline( 1/(1+beta_in)-1, ls='--', color='k', linewidth = 1, alpha=1)
    #plt.axvline( (-beta_in), ls='--', color='white', linewidth = 1, alpha=1)
    
    
for delta in [1/1.28]:
    plt.axvline( (1-1/delta), ls='--', color='white', linewidth = 1, alpha=1)
    #plt.axvline( (-beta_in), ls='--', color='white', linewidth = 1, alpha=1)
    
    
#Spec_rel
for delta in [1.28, 1.1, 1, 0.9]: #np.sqrt(1-0.000001*beta**2)
    plt.contour(X, Y, np.sqrt(1-beta**2)*1/(1+X), levels=[delta], linewidths = 3, linestyles='-', alpha=0.5, 
                color='purple')
    #plt.pcolormesh(X, Y, 1-np.sqrt(1-beta**2) / (1-X))

plt.ticklabel_format(style='plain') 
plt.xlabel(r'$\frac{z}{ct}$') 
plt.ylabel(r'$\frac{p}{ct}$') 
#plt.savefig('Pictures/Rel_resonance_1.png', dpi=300)
```

### $\tau$ scaling

```python
theta = np.linspace(0, 2*np.pi, 1000)
mu = np.cos(theta) #np.linspace(-1,1)
v = 0.1

plt.plot( mu, abs( (1-mu*v)**2/((1-v)*(mu*(mu-v)+(1-mu**2)*(1-v**2)))) )
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(10,10) )
for v in [0, 0.1, 0.2, 0.3]: 
    ax.plot(theta, abs( (1-mu*v)**2/((1-v)*(mu*(mu-v)+(1-mu**2)*(1-v**2)) ) ) ) 
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(10,10) )
for v in [0, 0.1, 0.2, 0.5]: 
    ax.plot(theta, abs( (1-mu*v)**2/(v*(1-v*mu)-mu*(1-v**2)) ) )
```
